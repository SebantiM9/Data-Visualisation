# **Bar Graph**

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color="black")
plt.show()
```

```python
### If I want to show the bars horizontally

x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color="green")
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,width=0.05)
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,height=0.1)
plt.show()
```

```python
y=np.array([10,90,50])

plt.pie(y)
plt.show()
```

```python
y=np.array([10,90,50])
mylabels=["Cherry","Apple","Mango"]
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([10,90,50])
mylabels=["Cherry","Apple","Mango"]
plt.pie(y,labels=mylabels,startangle=180)
plt.show()
```

```python
y=np.array([10,90,50])
mylabels=["Cherry","Apple","Mango"]
myexplode=[0.1,0,0.25]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```

```python
y=np.array([10,90,50])
mylabels=["Cherry","Apple","Mango"]
mycolors=["blue","r","yellow"]
plt.pie(y,labels=mylabels,startangle=180,colors=mycolors)
plt.show()
```
